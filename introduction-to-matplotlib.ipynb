## MatPlotLib example workFlow

```python
# 0. Import matplotlib and get it ready for plotting jupyer

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np




#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig,ax = plt.subplots(figsize=(5,5))  #(width,size)

#3. Add Plot data
ax.plot(x,y)

#4. Customize plots
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

plt.show()

#5. Save and show (you save the whole figure)
# fig.savefig("images/sample-plot.png")
# ax.show()
```

## Making figures with NumPy Array 

we want:
* Line plot
* Scatter plot
* Bar plot
* Histogram

```python
# Create some data

x = np.linspace(0,10, 100)
x[:10]
```

```python
# Plot the data and create the line plot
fig, ax = plt.subplots()
ax.plot(x, x ** 2);
plt.show()
```

```python
# Plot the data and create the scatter
fig,ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
```

```python
# Another scatter plot
fig,ax = plt.subplots()

ax.scatter(x, np.sin(x))
plt.show()
```

```python
# Make a plot from Dictionary

nut_butter_price = {
  "Almod butter": 10,
    "Peanut butter": 8,
    "Cashwe butter": 12
    
}

fig,ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values())
ax.set(title="HL Nut Butter Store",ylabel="price ($)");
plt.show();
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()), list(nut_butter_price.values()))
ax.set(title="HL Nut Butter Store", ylabel="price($)");
plt.show();
```

```python
# Make some data for histrogram and plot it
x = np.random.randn(1000)
x
fig, ax = plt.subplots()
ax.hist(x)
plt.show();
```

```python
## OO Method from Scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"], cmap='summer')

# Customize the plot
ax.set(title='Heart Diseases and Cholesterol Levels', xlabel="Age", ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--",color="red")

plt.show()
```

```python
# Subplot of chol, age , thalach
fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1,figsize=(10,10),sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"],cmap='winter')

# Customize the ax0
ax0.set(title='Heart Diseases and Cholesterol Levels', ylabel="Cholesterol")

# change the xaxis limit
ax0.set_xlim([50,80])

# add ax0 to a legend
ax0.legend(*scatter.legend_elements(), title="target")

# Add a horizontal line
ax0.axhline(over_50["chol"].mean(),linestyle="--",color='red')

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"],c=over_50["target"],cmap='winter')

# Customize the ax1
ax1.set(title='Heart Diseases and Max Heart Rate', xlabel="Age", ylabel="Max Heart Rate")
ax1.set_ylim([80,200])
# add ax1 to a legend
ax1.legend(*scatter.legend_elements(), title="target")

# Add a horizontal line
ax1.axhline(over_50["thalach"].mean(),linestyle="--",color='red')

# Add a title to the figure
fig.suptitle("Heart Diseases Analysis", fontsize=16, fontweight="bold");
# fig.suptitle("Heart Diseases Analysis", fontsize=16, fontweight="bold");


plt.show()
```
